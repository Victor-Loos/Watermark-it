Convertion de l'image

```python
from PIL import Image

def crop_image(img_path):
    img = Image.open(img_path)
    width, height = img.size

    if width > 2048:
        left = (width - 2048) // 2
        right = width - left
    else:
        left = 0
        right = width

    if height > 2048:
        top = (height - 2048) // 2
        bottom = height - top
    else:
        top = 0
        bottom = height

    img = img.crop((left, top, right, bottom))

    if img.size[0] > img.size[1]:
        diff = img.size[0] - img.size[1]
        left = diff // 2
        right = img.size[0] - (diff - left)
        img = img.crop((left, 0, right, img.size[1]))
    elif img.size[1] > img.size[0]:
        diff = img.size[1] - img.size[0]
        top = diff // 2
        bottom = img.size[1] - (diff - top)
        img = img.crop((0, top, img.size[0], bottom))

    plt.imshow(img)
    img.save(img_path+"crop.png")

crop_image("original/python.png")
```
